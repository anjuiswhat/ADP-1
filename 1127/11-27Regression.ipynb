```python
df = df.dropna()
```

```python
df['임대료'].value_counts()
```

```python
df[df['임대료'] == '-']
```

```python
df['임대료'] = df['임대료'].replace('-',0)
df['임대보증금'] = df['임대보증금'].replace('-',0)
```

```python
df['임대료'] = df['임대료'].astype(int)
df['임대보증금'] = df['임대보증금'].astype(int)
```

```python
df.columns
```

```python
df.info()
```

```python
col = ['단지코드', '총세대수', '임대건물구분', '지역', '공급유형', '공가수',
       '자격유형', '도보 10분거리 내 지하철역 수(환승노선 수 반영)',
       '도보 10분거리 내 버스정류장 수', '단지내주차면수', '등록차량수']
```

```python
aggregation = {
    '전용면적':'mean',
    '임대보증금':'mean',
    '임대료':'mean'
}
```

```python
df_gp = pd.DataFrame(df.groupby(col).agg(aggregation)).reset_index()
```

```python
df_gp
```

```python
df_gp.info()
```

```python
df_gp = df_gp.drop('자격유형', axis = 1)
```

```python
df_gp.describe()
```

```python
df_gp = df_gp.drop(['지역','임대건물구분','공급유형','단지코드'], axis = 1)
```

```python
X = df_gp[df_gp.columns.difference(['등록차량수'])]
y = df_gp['등록차량수']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV, KFold
```

```python
rf_param_grid = {
    'n_estimators': [100, 150, 200, 250],
    'max_depth': [None, 6, 9, 12],
    'min_samples_split': [0.01, 0.05, 0.1],
    'max_features': ['auto', 'sqrt'],
}
```

```python
rf = RandomForestRegressor()
kf = KFold(random_state = 0,
          n_splits = 10,
          shuffle = True)
```

```python
rf_grid_search = GridSearchCV(estimator = rf,
                          param_grid = rf_param_grid,
                          cv = kf,
                          n_jobs = -1,
                          verbose = 2)
```

```python
rf_grid_search.fit(X, y)
```

```python
rf_grid_search.best_params_
```
